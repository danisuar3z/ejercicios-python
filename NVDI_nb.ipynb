```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
band1 = np.load('Data/clip/LC08_L1TP_225084_20180213_20180222_01_T1_sr_band1_clip.npy')
band2 = np.load('Data/clip/LC08_L1TP_225084_20180213_20180222_01_T1_sr_band2_clip.npy')
band3 = np.load('Data/clip/LC08_L1TP_225084_20180213_20180222_01_T1_sr_band3_clip.npy')
band4 = np.load('Data/clip/LC08_L1TP_225084_20180213_20180222_01_T1_sr_band4_clip.npy')
band5 = np.load('Data/clip/LC08_L1TP_225084_20180213_20180222_01_T1_sr_band5_clip.npy')
band6 = np.load('Data/clip/LC08_L1TP_225084_20180213_20180222_01_T1_sr_band6_clip.npy')
band7 = np.load('Data/clip/LC08_L1TP_225084_20180213_20180222_01_T1_sr_band7_clip.npy')
bands = [band1, band2, band3, band4, band5, band6, band7]
```

```python
def crear_img_png(carpeta, n_band, band):
    low_q = np.quantile(band, 0.001)
    high_q = np.quantile(band, 0.999)
    plt.imshow(band, vmin=low_q, vmax=high_q)
    plt.colorbar(shrink=0.6)
    plt.savefig(os.path.join(carpeta, 'band' + str(n_band) + '.png'), dpi=1200)
    plt.show()

for n in range(1, 8):
    crear_img_png('Data/clip/pngs', n, bands[n-1])
```

```python
def crear_hist_png(carpeta, n_band, band, bins):
    plt.hist(band.flatten(), bins=bins, range=(np.quantile(band, 0.01), np.quantile(band, 0.99)))
    plt.savefig(os.path.join(carpeta, 'band' + str(n_band) + '_hist.png'))
    plt.show()

for n in range(1, 8):
    crear_hist_png('Data/clip/pngs', n, bands[n-1], 100)
```
